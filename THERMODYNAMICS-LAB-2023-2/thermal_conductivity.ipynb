```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
materiales = ["acrilico", "comprimido", "madera", "yeso", "vidrio"]
d1 = [9.4] * 5 #cm
d2 = {8.5:8.3, 8.4:8.4, 8.7:8.4, 8:8.3, 7.5:7.5} #cm
d2 = [abs(key + value) / 2 for key, value in d2.items()] #cm
ta = [10*60] * 5 #mins to secs
m_tot = [206.8, 208.19, 210.14, 208.84, 216.82] #g
m_vasito = [192.01]*5# + [161.67]*3
t = [5, 5, 5, 5, 2] #min
t = [value * 60 for value in t] #mins to secs
m_wa = [15.43] * 5 #g, ambiental
m_w = [] #g, m_tot - m_vasito

m_w = [m_tot[i] - m_vasito[i] for i in range(len(m_tot))]
```

```python
df = pd.DataFrame({"d1": d1, "d2": d2, "ta": ta, "m_tot": m_tot, "m_vaso": m_vasito, "t": t, "m_wa": m_wa, 
                   "m_w": m_w}, index = materiales)
df
```

```python
d_avg = [abs(d1[i] + d2[i]) / 2 for i in range(len(d1))]
A = [np.pi * (d_avg[i] / 2)**2 for i in range(len(d_avg))]
Ra = [m_wa[i]/ta[i] for i in range(len(m_wa))]
R = [m_w[i]/t[i] for i in range(len(m_w))]
Ro = [R[i] - Ra[i] for i in range(len(R))]

h = [0.5, 0.5, 0.4, 1, 0.5]

df["d_avg"] = d_avg
df["A"] = A
df["Ra"] = Ra
df["R"] = R
df["Ro"] = Ro
df["h"] = h
df
```

```python
# Cálculo de la conductividad térmica (k)
diferencial_temperatura = 100  # °C

df["k"] = (df["Ro"] * 80 * df["h"]) / (df["A"] * diferencial_temperatura)

# Visualizar el DataFrame con los resultados
df
```
