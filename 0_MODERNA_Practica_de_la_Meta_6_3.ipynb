a) El siguiente código grafica los estados de un pozo de potencial. Copia el código en jupyter y recupera las gráficas que se encuentran en esta página (wave functions) : https://towardsdatascience.com/quantum-physics-visualization-with-python-35df8b365ff

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Constants
h = 6.626e-34
m = 9.11e-31
```

```python
# Values for L and x
x_list = np.linspace(0, 1, 100)
L = 1
```

```python
def psi(n, L, x):
    return np.sqrt(2/L)*np.sin(n*np.pi*x/L)

def psi_2(n, L, x): # probabilidad
    return np.square(psi(n, L, x))
```

image.png

#GRAFICAS

```python
plt.figure(figsize=(15,10))
plt.suptitle("Wave Functions", fontsize=18)




for n in range(1, 4):
    # Empty lists for energy and psi wave
    psi_2_list = []
    psi_list = []
    
    for x in x_list:
        psi_2_list.append(psi_2(n, L, x))
        psi_list.append(psi(n, L, x))
    
    plt.subplot(3, 2, 2*n-1)
    plt.plot(x_list, psi_list)
    plt.xlabel("L", fontsize=13)
    plt.ylabel("Ψ", fontsize=13)
    plt.xticks(np.arange(0, 1, step=0.5))
    plt.title("n=" + str(n), fontsize=16)
    plt.grid()
    
    plt.subplot(3, 2, 2*n)
    plt.plot(x_list, psi_2_list)
    plt.xlabel("L", fontsize=13)
    plt.ylabel("Ψ*Ψ", fontsize=13)
    plt.xticks(np.arange(0, 1, step=0.5))
    plt.title("n=" + str(n), fontsize=16)
    plt.grid()




plt.tight_layout(rect=[0, 0.03, 1, 0.95])
```

## c) escribe un código para graficar la dependencia de los primeros 5 estados de energía con L, la longitud del pozo de potencial y otra con la masa.



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Constants
h = 6.626e-34
m = 9.11e-31
```

```python
# valores para L 
L_list = np.linspace(0.1, 1, 50)

# valores para m 
m_list = np.linspace(0.1*m, 10*m, 50) #desde una decima parte de la masa hasta 10 veces la masa
```

```python
# calcular energia_n
def energy_n(n, L, m):
  return (  (h**2 * np.pi**2) / (2 * m * L**2) ) * n**2
```

image.png

```python
# creamos subplots, dos filas y una columna 
fig, axs = plt.subplots(2, 1 ,figsize=(10, 12))


#energy vs L
for n in range(1, 6):
    energy_list = [energy_n(n, L, m) for L in L_list]
    axs[0].plot(L_list, energy_list, label=f'n = {n}')

axs[0].set_xlabel('L')
axs[0].set_ylabel('Energy')
axs[0].legend()

#energy vs m
for n in range(1, 6):
    energy_list = [energy_n(n, 1, mass) for mass in m_list]
    axs[1].plot(m_list, energy_list, label=f'n = {n}')
axs[1].set_xlabel('Mass')
axs[1].set_ylabel('Energy')
axs[1].legend()



# mostrar grafica
plt.show()
```
