```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
import math 
```

```python
y = np.array([2, 2.6, 4.7, 6.7]) #cm
m = np.array([1, 2, 3, 4]) #m
```

```python
df1 = pd.DataFrame({'y': y, 'm': m})
```

```python
#Realizar el ajuste lineal y mostrar los resultados
L_slope, L_intercept, L_r_value, L_p_value, L_std_err = linregress(df1['m'], df1['y'])

print('Pendiente:', L_slope)
print('Intersección:', L_intercept)
print('Coeficiente de correlación:', L_r_value)
print('Valor p:', L_p_value)
print('Error estándar:', L_std_err)
```

```python
plt.plot(df1['m'], df1['y'], 'bo')
plt.plot(df1['m'], L_slope*df1['m']+L_intercept, #y = mb + b
            'r-', label='Ajuste lineal')
plt.xlabel('m')
plt.ylabel('y')
plt.title('y vs m')
plt.show()
```
