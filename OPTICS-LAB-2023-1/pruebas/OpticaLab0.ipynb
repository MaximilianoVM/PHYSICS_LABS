```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
# Definir los datos
distancia = [15, 20, 25, 30, 35, 40, 45, 50, 55, 60]
lado = [3, 3.9, 4.6, 5.5, 6.4, 7.2, 8.2, 9, 10, 10.9]
```

```python
# Crear un dataframe de Pandas con los datos
df = pd.DataFrame({'distancia': distancia, 'lado': lado})

# Calcular el Ã¡rea (elevando el lado al cuadrado)
df['area'] = df['lado']**2

# Mostrar la tabla
print(df)
```

```python
# Graficar los datos con el lado
plt.plot(df['distancia'], df['lado'], 'bo')
plt.xlabel('Distancia')
plt.ylabel('Lado')
plt.title('Lado vs Distancia')
plt.show()
```

```python
# Graficar los datos con el Ã¡rea
plt.plot(df['distancia'], df['area'], 'bo')
plt.xlabel('Distancia')
plt.ylabel('Ãrea')
plt.title('Ãrea vs Distancia')
plt.show()
```

## LADO

```python
# LADO: Realizar el ajuste lineal y mostrar los resultados
L_slope, L_intercept, L_r_value, L_p_value, L_std_err = linregress(df['distancia'], df['lado'])

print('Pendiente:', L_slope)
print('IntersecciÃ³n:', L_intercept)
print('Coeficiente de correlaciÃ³n:', L_r_value)
print('Valor p:', L_p_value)
print('Error estÃ¡ndar:', L_std_err)
```

```python
# LADO: Graficar los datos con el ajuste lineal
plt.plot(df['distancia'], df['lado'], 
            'bo', label='Datos')

plt.plot(df['distancia'], L_slope*df['distancia']+L_intercept, #y = mb + b
            'r-', label='Ajuste lineal')
            
plt.xlabel('Distancia')
plt.ylabel('Lado')
plt.title('Lado vs Distancia con Ajuste Lineal')
plt.legend()
plt.show()
```

## AREA

```python
# AREA: Realizar el ajuste lineal y mostrar los resultados
A_slope, A_intercept, A_r_value, A_p_value, A_std_err = linregress(df['distancia'], df['area'])

print('Pendiente:', A_slope)
print('IntersecciÃ³n:', A_intercept)
print('Coeficiente de correlaciÃ³n:', A_r_value)
print('Valor p:', A_p_value)
print('Error estÃ¡ndar:', A_std_err)
```

```python
# Graficar los datos con el ajuste lineal
plt.plot(df['distancia'], df['area'], 
            'bo', label='Datos')
plt.plot(df['distancia'], A_slope*df['distancia']+A_intercept, #y = mb + b
            'r-', label='Ajuste lineal')
            
plt.xlabel('Distancia')
plt.ylabel('Ãrea')
plt.title('Ãrea vs Distancia con Ajuste Lineal')
plt.legend()
plt.show()
```
